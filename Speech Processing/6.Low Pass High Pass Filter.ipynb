```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import scipy.signal as sg
```

```python
d,sr = librosa.load('auds/set2.wav',sr=8000)
t = np.linspace(0,len(d)/sr,len(d))

b,a = sg.butter(4,2000/(sr/2),'high')
d_fil = sg.filtfilt(b,a,d)

b1,a1 = sg.butter(4,500/(sr/2),'low')
d1_fil = sg.filtfilt(b1,a1,d)

plt.figure(figsize=(14,7))
plt.plot(t,d)
plt.show()

fig,ax = plt.subplots(1,1,figsize=(14,7))
ax.plot(t,d)
ax.plot(t,d_fil,lw=1)
plt.show()

fig,ax = plt.subplots(1,1,figsize=(14,7))
ax.plot(t,d)
ax.plot(t,d1_fil,lw=1)
plt.show()
```
