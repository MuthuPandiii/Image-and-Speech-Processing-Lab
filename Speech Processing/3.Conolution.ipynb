```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
from scipy import signal
```

```python
d,sr = librosa.load('auds/set1.wav',sr=8000)
d1,sr1 = librosa.load('auds/set2.wav',sr=8000)
```

```python
y = signal.convolve(d,d1)
t3 = np.linspace(0,len(y)/sr,len(y))
```

```python
plt.subplot(221)
plt.plot(d)

plt.subplot(222)
plt.plot(d1)

plt.subplot(223)
plt.plot(t3,y)
```
