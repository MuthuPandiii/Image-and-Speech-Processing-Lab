# AutoCorrelation

```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
```

```python
d,sr = librosa.load('auds/sample1.wav',sr = 8000)
t = np.linspace(0,len(d)/sr,len(d))

y = librosa.autocorrelate(d,max_size=10000)

plt.figure(figsize=(14,7))

plt.subplot(211)
plt.plot(t,d)

plt.subplot(212)
plt.acorr(y)

plt.show()
```

# CrossCorrelation

```python
d,sr = librosa.load('auds/set1.wav',sr=8000)
t = np.linspace(0,len(d)/sr,len(d))

d1,sr = librosa.load('auds/set2.wav',sr=8000)
t1 = np.linspace(0,len(d)/sr,len(d))

y = np.correlate(d,d1,mode='full')
lags = np.arange(-100,100)

plt.figure(figsize=(14,7))
plt.subplot(311)
plt.plot(t,d)

plt.subplot(312)
plt.plot(t1,d1)

plt.subplot(313)
plt.plot(lags,y[0:200])
plt.grid()
plt.show()
```
