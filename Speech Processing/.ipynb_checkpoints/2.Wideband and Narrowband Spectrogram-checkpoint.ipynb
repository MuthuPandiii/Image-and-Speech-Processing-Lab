```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
import librosa.display
```

```python
y,sr = librosa.load('auds/speech.wav')

plt.figure(figsize=(14,7))
plt.plot(y)
plt.show()

fig,ax = plt.subplots()
mag1 = np.abs(librosa.core.stft(y,n_fft=1024,hop_length=10,win_length=40,window='hamming'))
mag1 = librosa.amplitude_to_db(mag1,ref = np.max)
img = librosa.display.specshow(mag1,sr=sr,x_axis='s',y_axis='linear')

ax.set_title('Wide Band')
fig.colorbar(img,ax=ax)

fig,ax = plt.subplots()
mag1 = np.abs(librosa.core.stft(y,n_fft=1024,hop_length=100,win_length=400,window='hamming'))
mag1 = librosa.amplitude_to_db(mag1,ref = np.max)
img = librosa.display.specshow(mag1,sr=sr,x_axis='s',y_axis='linear')

ax.set_title('Narrow Band')
fig.colorbar(img,ax=ax)
```
