```python
import wave
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
audio = wave.open('auds/sample1.wav')
params = audio.getparams()

frame_rate , n_frames = params[2:4]

data = np.frombuffer(audio.readframes(n_frames),dtype='int16')
t = np.linspace(0,len(data)/frame_rate,len(data))

def Energy_Calculation(data):
    sum_ = 0
    energy = []
    for i in range(len(data)):
        sum_ += int(data[i] ** 2)
        if (i + 1) % 256 == 0:
            energy.append(sum_)
            sum_ = 0
        if len(data) - 1 == i:
            energy.append(sum_)
    return energy

def Pitch(frame):
    return max(frame)

def ZCR(data,frame_size,overlap):
    wlen = len(data)
    step = frame_size - overlap
    frames = math.ceil(wlen / step)
    zcr = np.zeros((frames,1))
    for i  in range(frames):
        curFrame = data[np.arange(i*step,min(i * step + frame_size,wlen))]
        curFrame = curFrame - np.mean(curFrame)
        zcr[i] = sum(curFrame[0:-1]*curFrame[1::]<=0)
    return zcr


Energy = Energy_Calculation(data)

fsize = 1024
nframes = int(len(data)/fsize)
frequencies = np.zeros((nframes))
for i in range(nframes):
    frame = data[i * fsize : (i+1)*fsize]
    frequencies[i] = Pitch(frame)
    
zcr = ZCR(data,fsize,overlap=512)

plt.figure(figsize=(14,7))
plt.subplot(221)
plt.plot(t,data)

plt.subplot(222)
plt.plot(Energy)

plt.subplot(223)
plt.plot(frequencies)

plt.subplot(224)
plt.plot(zcr)

plt.show()
```
