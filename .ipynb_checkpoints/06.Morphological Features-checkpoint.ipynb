```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.cvtColor(cv.imread('C:\\Users\\HP\\OneDrive\\Documents\\Lab\\Image and Speech Processing Lab\\Images\\image 3.jpg'),cv.COLOR_BGR2RGB)
```

```python
filter_ = np.ones((5,5),np.uint8)
```

```python
eroded_img = cv.erode(img,filter_,iterations=1)
dilated_img = cv.dilate(img,filter_,iterations = 1)
```

```python
plt.figure(figsize=(14,7))
plt.subplot(131)
plt.imshow(img)
plt.subplot(132)
plt.imshow(eroded_img)
plt.subplot(133)
plt.imshow(dilated_img)
```
