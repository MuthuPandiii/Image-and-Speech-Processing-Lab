```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
img =cv.cvtColor(cv.imread('C:\\Users\\HP\\OneDrive\\Documents\\Lab\\Image and Speech Processing Lab\\Images\\image 3.jpg'),cv.COLOR_BGR2RGB)
```

```python
print(img.shape)
```

```python
input_data = np.float32(img.reshape((-1,3)))
```

```python
input_data.shape
```

```python
criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER,100,0.85)
```

```python
retval,labels,centers = cv.kmeans(input_data,3,None,criteria,10,cv.KMEANS_RANDOM_CENTERS)
```

```python
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]
```

```python
segmented_img = segmented_data.reshape(img.shape)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(121)
plt.imshow(img)
plt.subplot(122)
plt.imshow(segmented_img)
plt.show()
```
