```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('C:\\Users\\HP\\OneDrive\\Documents\\Lab\\Image and Speech Processing Lab\\Images\\image 3.jpg')
```

```python
vertical_filter = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
horizontal_filter = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
```

```python
vertical_edge = cv.filter2D(img,-1,vertical_filter)
horizontal_edge = cv.filter2D(img,-1,horizontal_filter)
```

```python
fig,axes = plt.subplots(1,2,figsize=(15,15))
axes[0].imshow(vertical_edge)
axes[0].set_title("Vertical Edges")
axes[1].imshow(horizontal_edge)
axes[1].set_title("Horizontal Edges")
plt.show()
```

## Canny Edge Detection

```python
img = cv.imread('C:\\Users\\HP\\OneDrive\\Documents\\Lab\\Image and Speech Processing Lab\\Images\\image 3.jpg',-1)
```

```python
canny = cv.Canny(img,100,200)
```

```python
fig,axes = plt.subplots(1,2,figsize=(15,15))
axes[0].imshow(img)
axes[0].set_title("Original Image")
axes[1].imshow(canny,cmap='gray')
axes[1].set_title("Canny Edges")
plt.show()
```
