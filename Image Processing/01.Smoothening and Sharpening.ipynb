```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.cvtColor(cv2.imread('./Images/image 1.jpg'),cv2.COLOR_BGR2RGB)

smoothened_img = cv2.blur(img,(3,3))

kernel = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
sharpened_img = cv2.filter2D(img,-1,kernel)

plt.figure(figsize=(20,20))

plt.subplot(311)
plt.imshow(img)
plt.title('Original Image')
plt.xticks([])
plt.yticks([])

plt.subplot(312)
plt.imshow(smoothened_img)
plt.title('Smoothened Image')
plt.xticks([])
plt.yticks([])

plt.subplot(313)
plt.imshow(sharpened_img)
plt.title('Sharpened Image')
plt.xticks([])
plt.yticks([])
plt.show()
```
