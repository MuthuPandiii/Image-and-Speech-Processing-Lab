```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.cvtColor(cv2.imread('./Images/image 3.jpg'),cv2.COLOR_BGR2RGB)

h_filter = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])

v_filter = np.flip(h_filter.T)

v_edge = cv2.filter2D(img,-1,v_filter)
h_edge = cv2.filter2D(img,-1,h_filter)
```

```python
plt.figure(figsize=(20,20))

plt.subplot(311)
plt.imshow(img)
plt.title('Original Image')
plt.xticks([])
plt.yticks([])

plt.subplot(312)
plt.imshow(h_edge)
plt.title('Horizontal Edges')
plt.xticks([])
plt.yticks([])

plt.subplot(313)
plt.imshow(v_edge)
plt.title('Vertical Images')
plt.xticks([])
plt.yticks([])
plt.show()
```

## Canny  Edge Detection

```python
img = cv2.imread('./Images/image 3.jpg',0)

canny_edge = cv2.Canny(img,100,200)

plt.figure(figsize=(14,7))
plt.imshow(canny_edge)
plt.show()
```
