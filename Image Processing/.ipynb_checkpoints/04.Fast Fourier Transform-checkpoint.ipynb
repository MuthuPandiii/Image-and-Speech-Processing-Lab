```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('./Images/image 2.jpg',0)
```

```python
img1 = np.fft.fft2(img)
img2 = np.fft.fftshift(img1)
img3 = np.fft.ifftshift(img2)
img4 = np.fft.ifft2(img3)

plt.figure(figsize=(20,20))

plt.subplot(511)
plt.imshow(img,'gray')

plt.subplot(512)
plt.imshow(np.log(1 + np.abs(img1)),'gray')

plt.subplot(513)
plt.imshow(np.log(1 + np.abs(img2)),'gray')

plt.subplot(514)
plt.imshow(np.log(1 + np.abs(img3)),'gray')

plt.subplot(515)
plt.imshow(np.log(1 + np.abs(img4)),'gray')

plt.show()
```
