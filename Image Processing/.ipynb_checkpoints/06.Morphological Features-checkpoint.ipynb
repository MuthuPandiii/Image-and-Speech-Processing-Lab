```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.cvtColor(cv.imread('./Images/image 2.jpg',0),cv.COLOR_BGR2RGB)
```

```python
kernel = np.ones((5,5),np.uint8)

erode = cv.erode(img,kernel,iterations=1)
dilate = cv.dilate(img,kernel,iterations = 1)

plt.figure(figsize=(20,20))

plt.subplot(311)
plt.imshow(img)

plt.subplot(312)
plt.imshow(erode)

plt.subplot(313)
plt.imshow(dilate)
```
