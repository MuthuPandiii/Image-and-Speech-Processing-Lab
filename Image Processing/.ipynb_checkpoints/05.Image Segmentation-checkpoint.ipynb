```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.cvtColor(cv.imread('./Images/image 2.jpg'),cv.COLOR_BGR2RGB)
```

```python
pixel_vals = np.float32(img.reshape((-1,3)))

criteria = (cv.TERM_CRITERIA_EPS+cv.TERM_CRITERIA_MAX_ITER,100,0.85)

_,labels,centers = cv.kmeans(pixel_vals,3,None,criteria,10,cv.KMEANS_RANDOM_CENTERS)

centers = np.int8(centers)

segmented_data = np.abs(centers[labels.flatten()])
segmented_data = segmented_data.reshape(img.shape)

plt.figure(figsize=(20,20))

plt.subplot(211)
plt.imshow(img)

plt.subplot(212)
plt.imshow(segmented_data)

plt.show()
```
